In this exercise, we will import, manipulate, and visualize two datasets of star clusters in our Galaxy.  Before you begin, make sure you have the files GlobularClusters.tab and OpenClusters.tab in your data directory.

An **open cluster** is a loosely-associated cluster of hundreds to thousands of stars.  They are usually young-ish stars which formed from the same birth cloud.  These clusters are likely to be disrupted by gravitational forces within the Milky Way.

A **globular cluster** is a gravitationally-bound cluster of hundreds of thousands of stars.  Because the cluster is self-gravitating, it can survive intact and usually contains very old stars.

You will see that the table contains the following fields:

## Visualizing the data

Make a scatter plot of apparent visual magnitude as a function of distance from the Sun for the set of open clusters.  Make sure to: 
- Label your axes, and 
- Turn the y-axis upside down so that magnitude decreases (brightness increases) going up.

```python
# Open clusters scatter plot
plt.scatter(open_clusters['distSun(kly)'], open_clusters['m_v'], c='blue', label='Open Clusters')
plt.xlabel('Distance from Sun (kly)')
plt.ylabel('Apparent Magnitude (m_v)')
plt.gca().invert_yaxis()  # Invert y-axis for magnitude
plt.title('Open Clusters: Apparent Magnitude vs Distance')
plt.show()
# YOUR CODE HERE
```

Now make the same plot for the globular clusters.

```python
# Globular clusters scatter plot
plt.scatter(globular_clusters['distSun(kly)'], globular_clusters['m_v'], c='green', label='Globular Clusters')
plt.xlabel('Distance from Sun (kly)')
plt.ylabel('Apparent Magnitude (m_v)')
plt.gca().invert_yaxis()  # Invert y-axis for magnitude
plt.title('Globular Clusters: Apparent Magnitude vs Distance')
plt.show()
# YOUR CODE HERE
```

Now combine them, putting the open and globular clusters on the same plot.  Make sure to: 
- Give the two datasets different colors to differentiate them, 
- Include a legend, and
- Set the axis limits to concentrate on where most of the data is.

```python
# Combined scatter plot
plt.scatter(open_clusters['distSun(kly)'], open_clusters['m_v'], c='blue', label='Open Clusters', alpha=0.6)
plt.scatter(globular_clusters['distSun(kly)'], globular_clusters['m_v'], c='green', label='Globular Clusters', alpha=0.6)
plt.xlabel('Distance from Sun (kly)')
plt.ylabel('Apparent Magnitude (m_v)')
plt.gca().invert_yaxis()  # Invert y-axis for magnitude
plt.legend()
plt.xlim(0, 50)  # Adjust limits for clarity
plt.ylim(-5, 20)
plt.title('Apparent Magnitude vs Distance for Star Clusters')
plt.show()
# YOUR CODE HERE
```

Here we can see that the globular and open cluster samples are at very different distances, despite having similar apparent brightnesses!  This makes sense given what we know about the difference between open and globular clusters - globular clusters have at least 100 times as many stars, so their *intrinsic* brightness should be higher, which means they must be further away.

Now let's see how the two samples are distributed on the sky.

Plot the coordinates (RA and DEC) of the two samples, keeping in mind all the plotting best practices we have mentioned so far.

```python
# Plot coordinates
plt.scatter(open_clusters['RA'], open_clusters['DEC'], c='blue', label='Open Clusters', alpha=0.6)
plt.scatter(globular_clusters['RA'], globular_clusters['DEC'], c='green', label='Globular Clusters', alpha=0.6)
plt.xlabel('Right Ascension (RA)')
plt.ylabel('Declination (DEC)')
plt.legend()
plt.title('Sky Distribution of Star Clusters')
plt.show()
# YOUR CODE HERE
```

So, not only are the two sets of clusters at different distances, but they are distributed differently on the sky!  The peculiar path traced out by the open clusters represents the plane of the galactic disk in these coordinates.  The **disk** is where most of the recent and ongoing star formation is happening in the galaxy.  The globular clusters, meanwhile, are clustered around one spot.  The coordinates of that spot mark the Galactic Center, indicating that most of the globular clusters are in the **bulge** of old stars around the center of the galaxy.
