## Slicing and reshaping arrays

## Using 1D arrays to create 2D maps

In the following exercise we will create maps (i.e., 2D arrays) of some function $z = f(x,y)$.  First, let's create our x- and y- axes.

Use np.linspace() to create an array x varying from 0 to 100, and an array y varying from -50 to 50, with 101 elements each.

```python
x = np.linspace(0, 100, 101)
y = np.linspace(-50, 50, 101) # YOUR CODE HERE
```

The numpy function **meshgrid()**, below, creates 2D arrays of x and y from those 1D arrays.  Print out the first two rows of xx and the first two rows of yy.

```python
xx, yy = np.meshgrid(x,y)
print(xx[:2,:])
print(yy[:2,:])
```

Print out the first 10 columns of xx and the first 10 columns of yy.

```python
print(xx[:,:11])
print(yy[:,:11])# YOUR CODE HERE
```

As you can see, the array xx has identical *rows* which are all equal to the array x, and the array yy has identical *columns* which are all equal to the array y.  Thus, xx gives the x-location of every spot in the map, while yy gives the y-location of every spot in the map.  Every combination of indices $[i,j]$ will specify a different location in the xy plane.

Visualize the array xx using the imshow() function.  Make sure to include a color bar.

```python
plt.imshow(xx)
plt.colorbar() # YOUR CODE HERE
```

Now do the same for the array yy.

```python
plt.imshow(yy)
plt.colorbar()# YOUR CODE HERE
```

Notice that the color bars of the two arrays are different, because the xx and yy arrays span different values.  Also notice that the lowest values of $y$ are at the top of the image, which is not what we usually want.  Make another version of the yy map with the origin in the bottom left corner (and continue using this convention for the rest of the exercise) using the *origin* keyword argument.

```python
plt.imshow(yy, origin = 'lower')
plt.colorbar()# YOUR CODE HERE
```

Below I have defined two variables $a$ and $b$.  Visualize the function $f(x,y) = a*x + b*y$. Experiment with changing the values of $a$ and $b$ (including making them negative) to see how the slope of the gradient changes.

Visualize the function $r = \sqrt{x^2 + y^2}$ showing distance from the origin (0,0).

```python
r = np.sqrt(xx**2 + yy**2)
plt.imshow(r, origin = 'lower') 
plt.colorbar() # YOUR CODE HERE
```

The lowest point should be located at (0,0), but the axes don't reflect that!  Update the axis ranges using the *extent* keyword argument to reflect the ranges of x and y.

```python
r = np.sqrt(xx**2 + yy**2)
plt.imshow(r, origin = 'lower', extent = (0, 100, -50, 50)) 
plt.colorbar() # YOUR CODE HERE
```

Calculate and print the number of pixels in this map which have a radius value $r < 10$.  (Hint: you can create a *Boolean mask* to select only the pixels which satisfy a certain condition with 2d arrays just like you can with 1d arrays!) 

```python
r_less_10 = r < 10
print(len(r[r_less_10]))
```

Now do the same for 
$$10 < r < 20$$
$$20 < r < 30$$
$$30 < r < 40$$
$$40 < r < 50$$

Calculate and print out the number of pixels which lie in each ring.

```python
r_less_20 = (r > 10) & (r < 20)
print(len(r[r_less_20]))
r_less_30 = (r > 20) & (r < 30)
print(len(r[r_less_30]))
r_less_40 = (r > 30) & (r < 40)
print(len(r[r_less_40]))
r_less_50 = (r > 40) & (r < 50)
print(len(r[r_less_50]))
```

If you have done this correctly, you should find that your count approximates
$$\pi~r~dr,$$
that is, half the area of a ring at that radius $r$ with that width $dr$!  I have calculated and printed this below for you to check your work against.

```python
from math import pi
r_mid = np.array([5,15,25,35,45])
dr = 10.
print(pi*r_mid*dr)
```
